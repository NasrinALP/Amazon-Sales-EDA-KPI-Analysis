### Data Cleaning

üîò removing useless columns

üîò removing duplicates

üîò Changing Data Types

üîò Populating Null values (in rating_count column)

üîò removing currency and percentage signs

üîò removing duplicates



```python
# removing signs: '‚Çπ' and ',' and '%' 
df['discounted_price'] = df['discounted_price'].str.replace('‚Çπ','')
df['discounted_price'] = df['discounted_price'].str.replace(',','')

df['actual_price'] = df['actual_price'].str.replace("‚Çπ",'')
df['actual_price'] = df['actual_price'].str.replace(",",'')

df['discount_percentage'] = df['discount_percentage'].str.replace('%','')

df['rating_count'] = df['rating_count'].str.replace(',', '')
```

```python
# changing data types from object to float
df['discounted_price'] = df['discounted_price'].astype('float64')
df['actual_price'] = df['actual_price'].astype('float64')
df['discount_percentage'] = df['discount_percentage'].astype('float64')
df['rating_count'] = df['rating_count'].astype('float64')
```

```python
# mapping to 0-1
df['discount_percentage'] = df['discount_percentage'] / 100
```

```python
# filling missing value with median in rating column
df['rating'] = df['rating'].str.replace('|','4.1')
df['rating'] = df['rating'].astype('float64')
```

```python
# filling missing values with median in rating_count column
df['rating_count'] = df.rating_count.fillna(value=df['rating_count'].median())
```

### Q1

```python
# Select relevant columns
corr_features = ['discount_percentage', 'discounted_price', 'rating_count']

# Compute correlation matrix
corr_matrix = df[corr_features].corr()

# Display correlation heatmap
plt.figure(figsize=(6, 4))
sb.heatmap(corr_matrix, annot=True, fmt=".2f")
plt.title('Correlation: Discount %, Discounted Price, and Rating Count')
plt.show()

# Optional KPI: Price Sensitivity Index
price_sensitivity = corr_matrix.loc['discount_percentage', 'rating_count']
print(f"Price Sensitivity Index (corr between discount% and rating_count): {price_sensitivity:.3f}")
```

### Q2

```python
# split category path for better insights
```

```python
df['category_levels'] = df['category'].str.split('|').apply(len)
df['category_levels'].min()
```

```python
df['Category'] = df['category'].str.split('|').str[1]
```

```python
np.shape(df['category'].str.split('|').str[1].unique())
```

```python
# Weighted average rating per category
category_satisfaction = (
    df.groupby('Category')
      .apply(lambda x: (x['rating'] * x['rating_count']).sum() / x['rating_count'].sum())
      .reset_index(name='weighted_avg_rating')
      .sort_values('weighted_avg_rating', ascending=False)
)

print(category_satisfaction.head())

# Visualization
plt.figure(figsize=(10, 5))
sb.barplot(data=category_satisfaction, x='Category', y='weighted_avg_rating')
plt.xticks(rotation=45, ha='right')
plt.title('Weighted Average Rating by Category')
plt.ylabel('Weighted Average Rating')
plt.xlabel('Category')
plt.show()
```

### Q3

```python
# Correlation between discount_percentage and rating
corr_discount_rating = df[['discount_percentage', 'rating']].corr().iloc[0, 1]
print(f"Correlation between discount percentage and rating: {corr_discount_rating:.3f}")

# Scatter plot visualization
plt.figure(figsize=(6, 4))
sb.scatterplot(data=df, x='discount_percentage', y='rating', alpha=0.6)
sb.regplot(data=df, x='discount_percentage', y='rating', scatter=False, color='red')
plt.title('Discount Percentage vs Rating')
plt.show()
```

### Q4

```python
# Avoid division by zero
df_filtered = df[df['discounted_price'] > 0].copy()
df_filtered['engagement_to_price'] = df_filtered['rating_count'] / df_filtered['discounted_price']

# Average by category
engagement_kpi = (
    df_filtered.groupby('Category')['engagement_to_price']
    .mean()
    .reset_index()
    .sort_values('engagement_to_price', ascending=False)
)

print(engagement_kpi.head())

# Visualization
plt.figure(figsize=(10, 5))
sb.barplot(data=engagement_kpi, x='Category', y='engagement_to_price')
plt.xticks(rotation=45, ha='right')
plt.title('Engagement-to-Price Ratio by Category')
plt.ylabel('Engagement-to-Price Ratio')
plt.xlabel('Category')
plt.show()
```

### Q5

```python
# Select numeric columns for correlation analysis
numeric_df = df.select_dtypes(include='number')

# Compute correlation matrix
corr_all = numeric_df.corr()

# Display top correlated features with rating and rating_count
rating_corr = corr_all['rating'].sort_values(ascending=False)
rating_count_corr = corr_all['rating_count'].sort_values(ascending=False)

print("Top correlations with Rating:")
print(rating_corr.head())

print("\nTop correlations with Rating Count:")
print(rating_count_corr.head())

# Visualize correlation heatmap
plt.figure(figsize=(8, 6))
sb.heatmap(corr_all, annot=True)
plt.title('Correlation Heatmap (All Numeric Features)')
plt.show()
```
